## Project: PDF creator to payroll

```python
import pandas as pd
import numpy as np
import pdfkit
import os
from generate_html import * 

# Config of wkhtmltopdf
path_to_wkhtmltopdf = r'C:\Program Files\wkhtmltopdf\bin\wkhtmltopdf.exe'
config = pdfkit.configuration(wkhtmltopdf=path_to_wkhtmltopdf)

# dataframe
data = pd.read_csv('data_company_IT.csv')  
data
```

```python
df = data.iloc[0:1] # example , replace for real dataframe


# create a folder 
output_folder = "Employees_pdf"
os.makedirs(output_folder, exist_ok=True)

logo_path = os.path.abspath("fake_company.png")
footer1_path = os.path.abspath("fake_signature.png")
footer2_path = os.path.abspath("worker_signature.png")
logo_url = f"file:///{logo_path.replace(os.sep, '/')}"
footer1_url = f"file:///{footer1_path.replace(os.sep, '/')}"
footer2_url = f"file:///{footer2_path.replace(os.sep, '/')}"

options = {
    'enable-local-file-access': None
}

for index, row in df.iterrows():
    html_content = generate_html_v2(row, logo_url, footer1_url,footer2_url) #change HTML version
    output_pdf_path = os.path.join(output_folder, f"{row['First Name']}_{row['Last Name']}_payment.pdf")
    pdfkit.from_string(html_content, output_pdf_path, configuration=config, options=options)
    print(f"{row['First Name']}_{row['Last Name']}_payment.pdf")
```
